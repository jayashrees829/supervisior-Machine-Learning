Task no.1:-Predict the percentage of marks of an student based on the nuber of study hours

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url= 'http://bit.ly/w-data'
dataset= pd.read_csv(url)
dataset.head()
```

```python
dataset.info()
```

```python
sns.pairplot(dataset)
```

```python
x= dataset.iloc[:,0:1]
x.head()
```

```python
y= dataset.iloc[:,1:]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(x_train, y_train)
predictions= model.predict(x_test)
```

```python
comparison= pd.DataFrame(np.c_[y_test, predictions],columns=['Original Score','Predictions'])
comparison
```

```python
from sklearn import metrics
print('MSE:',metrics.mean_squared_error(y_test, predictions))
print('MAE:',metrics.mean_absolute_error(y_test, predictions))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
fig= plt.figure()
axes= fig.add_axes([0,0,1,1])
axes.scatter(x,y)
axes.plot(x_test,predictions,color='red')
```

```python
A= model.predict([[9.5]])
print('If student studies 9.5 hrs/day he would get {} percentage.'.format(A[0][0]))
```
